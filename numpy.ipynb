```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:\n", arr)
print("Shape:", arr.shape)  # Output: (2, 3)
print("Number of dimensions:", arr.ndim)  # Output: 2
print("Size (number of elements):", arr.size)  # Output: 6
print("Data type:", arr.dtype)  # Output: int32 (may vary based on platform)
print("Item size (in bytes):", arr.itemsize)  # Output: 8 (may vary based on platform)
```

Numpy Vectorized Operation

```python
arr1=np.array([1,2,3,4,5])
arr2=np.array([10,20,30,40,50])

#operations
addition = arr1 + arr2
substraction = arr2 - arr1
multiplication = arr1*arr2
division = arr1/arr2

print(addition)
print(substraction)
print(multiplication)
print(division)
```

```python
#Universal Functions

arr1=(1,2,3,4,5)

print(np.sqrt(arr1))
print(np.sin(arr1))
print(np.exp(arr1))
print(np.log(arr1))
```

Some practical use of arrays

```python
# Stastical concept -------- NORMALIZATION
# It means mean of 0 and standard deviation of 1

data = np.array([1,2,3,4,5])

# Calculation mean and SD
mean = np.mean(data)
std_dev = np.std(data)

# Normalization = (data-mean)/sd

normalization = (data-mean)/std_dev

print("Normalized data =", normalization)
```

Calculating Mean, median, sd, variance

```python
arr= np.array([1,2,3,4,5])

mean = np.mean(arr)
median = np.median(arr)
sd = np.std(arr)
variance = np.var(arr)

print(mean)
print(median)
print(sd)
print(variance)
```
