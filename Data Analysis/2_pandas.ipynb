A series is one dimensional array like object that can hold any data type.
It is similar to a column in a table.

```python
import pandas as pd

# Creating a series

data = (1,2,3,4,5)

series = pd.Series(data)

print(series)
```

```python
#creating a series using dictionary

dictn = {'a': 1, 'b': 2, 'c': 3}

series_dict = pd.Series(dictn)

print(series_dict)
```

```python
#medhod 3

data = [1,2,3,4]
ind = ["a","b","c","d"]

series_3 = pd.Series(data, index=ind)

print(series_3)
```

Dataframes

```python
data = {
    "Name" : ['Satyam','Shivam','Sundaram'],
    "City" : ['New Yory','London','Bihta'],
    "Age" : [42,43,24]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Creating DataFrames from Dictionary

data = [
    {'Name':'Satyam','Age': 16,'City':'Delhi'},
    {'Name':'Shivam','Age': 18,'City':'Patna'},
    {'Name':'Sundaram','Age': 24,'City':'Bihta'}
]

df = pd.DataFrame(data)

print(df)
```

```python
df.iloc[0][2]
```

```python
#Accessing a specific element

df.at[2,"Name"]
```

```python
df.iat[2,2]
```
