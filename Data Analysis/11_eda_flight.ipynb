```python
#Processing Date

df['Date']= df["Date_of_Journey"].str.split('/').str[0]
df['Month']= df["Date_of_Journey"].str.split('/').str[1]
df['Year']= df["Date_of_Journey"].str.split('/').str[2]

df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop("Date_of_Journey", axis = 1, inplace=True)
```

```python
# Processing Arrivcal Time

df['Arrival_Hour'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(' ').str[0].str.split(':').str[1]

df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)

df.drop("Arrival_Time", axis = 1, inplace=True)

df.head()
```

```python
# Processing Departure Time

df['Dep_Hour']= df["Dep_Time"].str.split(':').str[0]
df['Dep_Minute']= df["Dep_Time"].str.split(':').str[1]

df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Minute'] = df['Dep_Minute'].astype(int)

df.drop("Dep_Time", axis = 1, inplace=True)

df.head()
```

```python
df.info()
```

```python
df.Total_Stops.unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna('non-stop')

stop_map = {
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4,
}

df['Total_Stops'] = df['Total_Stops'].map(stop_map)

df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
df.drop("Route", axis = 1, inplace=True)
```

```python
duration_hours = []
duration_mins = []

for duration in df['Duration']:
    parts = duration.split()
    
    if len(parts) == 2:
        duration_hours.append(int(parts[0][:-1]))
        duration_mins.append(int(parts[1][:-1]))
        
    else:
        if 'h' in parts[0]:
            duration_hours.append(int(parts[0][:-1]))
            duration_mins.append(0)
        else:
            duration_hours.append(0)
            duration_mins.append(int(parts[0][:-1]))



df['Duration_Hour'] = duration_hours
df['Duration_Minute'] = duration_mins


df.drop("Duration", axis=1, inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

encoded_data = encoder.fit_transform(df[['Airline', 'Source', 'Destination']])

encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out())

df = pd.concat([df, encoded_df], axis=1)
df.drop(['Airline', 'Source', 'Destination'], axis=1, inplace=True)

df.head()
```
